```python
import re
import pandas as pd
import numpy as np
import nltk
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
```

```python
data = pd.read_csv("Combined_data.csv")
```

```python
data["Cleantext"]= data["text"].str.lower()
stop = stopwords.words('english')
data['Cleantext'] = data['Cleantext'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
data['Cleantext'] = data['Cleantext'].str.replace('\d+', '')
data["Cleantext"] = data['Cleantext'].str.replace('[^\w\s]','')
```

```python
data.head()
```

```python
def tokenize(txt):
    tokens = re.split('\W+',txt)
    return tokens
data['tokenized']= data['Cleantext'].apply(lambda x: tokenize(x.lower()))
```

```python
wn=nltk.WordNetLemmatizer()
ps=nltk.PorterStemmer()
dir(wn)
dir(ps)
```

```python
def lemmatization(txt):
    ttxt =[wn.lemmatize(word) for word in txt ]
    return ttxt
```

```python
data['lemmatized']=data['tokenized'].apply(lambda x: lemmatization(x))
data.head()
```

```python
def stemming(txt):
    ttxt =[ps.stem(word) for word in txt ]
    return ttxt
```

```python
data['stemtext']=data['lemmatized'].apply(lambda x: stemming(x))
data.head()
```
