```python
import pandas as pd 
import xml.etree.ElementTree as et 
import os
```

```python
df_cols = ["text","labels"]
rows = []
for i in os.listdir("xml"):
    i= './xml/'+i
    xtree = et.parse(i)
    xroot = xtree.getroot() 
    for node in xroot: 
        s_body = node.find("body").text
        rows.append({"text": s_body,"label" : "1"})
        out_df =pd.DataFrame(rows, columns = df_cols)    
```

```python
out_df['labels']=0
```

```python
out_df.drop_duplicates()
```
